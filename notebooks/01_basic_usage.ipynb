# Belief-Agent v2: Basic Usage

This notebook walks through the core data model and agent wrapper.

```python
import json, sys
sys.path.insert(0, "..")
from belief_agent import BeliefAgent, BeliefState
```

## 1. BeliefState

A belief has a proposition, confidence score, evidence list, and contradictions.

```python
b = BeliefState(
    proposition="LLMs can reason",
    confidence=0.75,
    evidence=["Chain-of-thought improves accuracy"],
)
print(b)
print(f"Confident? {b.is_confident()}")
print(f"Contradicted? {b.is_contradicted()}")
```

### Merge beliefs

```python
b1 = BeliefState("Python is fast", confidence=0.8, evidence=["Benchmarks"])
b2 = BeliefState("Python is fast", confidence=0.6, contradictions=["GIL limits speed"])
merged = b1.merge(b2)
print(f"Merged confidence: {merged.confidence:.2f}")
print(f"Evidence: {merged.evidence}")
print(f"Contradictions: {merged.contradictions}")
```

## 2. BeliefAgent

Wrap any LLM with a `llm_call` function.

```python
def fake_llm(messages):
    return json.dumps({"proposition": "AI is evolving", "confidence": 0.8})

agent = BeliefAgent(llm_call=fake_llm, name="demo", system_prompt="You are helpful.")
agent.adopt(BeliefState("The sky is blue", confidence=0.95))
print(f"Beliefs: {len(agent.beliefs)}")
print(agent._belief_context())
```

```python
reply = agent("What's new?")
print(f"Agent: {reply}")
print(f"Beliefs now: {len(agent.beliefs)}")
for b in agent.list_beliefs():
    print(f"  - {b}")
```

## 3. Serialization

```python
data = agent.to_dict()
print(json.dumps(data, indent=2))
```
